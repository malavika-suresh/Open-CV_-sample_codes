```
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```
image = cv2.imread('/Sign_board/data/sign1.jpg')
imgrgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
img = imgrgb
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blur = cv2.medianBlur(gray,5)
cimg = cv2.cvtColor(blur,cv2.COLOR_GRAY2BGR)

circles = cv2.HoughCircles(blur,cv2.HOUGH_GRADIENT,1,200,param1=100,param2=40,minRadius=45,maxRadius=0)
circles = np.uint16(np.around(circles))
for  c in circles[0,:]:
    
    cv2.circle(cimg,(c[0],c[1]),c[2],(0,255,0),2)
    c = c.astype(int)
    # get the actual cropped images here
    crop = img[c[1]-c[2]:c[1]+c[2], c[0]-c[2]:c[0]+c[2]]
plt.imshow(cimg)
plt.show()
plt.imshow(crop)
plt.show()
```

```
cv2.imwrite('sign1.jpeg',crop )
```
